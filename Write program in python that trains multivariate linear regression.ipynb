```python
#importing numpy package
import numpy as np
DATA = 100
epoch = 150
np.random.seed(18)
```

```python
#creating x1, x2, x3, variables
x = np.random.rand(DATA,1)
x1 = x**1 + np.random.rand(DATA,1)**(1)
x2 = x**2 + np.random.rand(DATA,1)**(2)
x3 = x**3 + np.random.rand(DATA,1)**(3)
```

```python
#creating dependent variable alog with independent
y = 2*x1  + 5 + np.random.rand(DATA,1) + 5*x2 + 3*x3

#plotting the data
import matplotlib.pyplot as mpy
mpy.scatter(x1,y)
mpy.show()
```

```python
#scaling the data using mean and standard deviation
def scalerFunction(x):
    x = (x - np.mean(x))/np.std(x)
    return x
```

```python
x1= scalerFunction(x1)
x2= scalerFunction(x2)
x3= scalerFunction(x3)

#multiple linear regression function
def linearRegression(w1,w2,w3,b):
    return w1*x1 +b + w2*x2 + w3*x3
```

```python
#loss function
def MSE(yhat,y):
    return np.sum((yhat-y)**2)/len(y)
```
